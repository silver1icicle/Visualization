# Scatter plots using Bokeh
## Bokeh(BokehJS) is an interactive visualization library that targets modern web browsers for presentation. Its goal is to provide elegant, concise construction of versatile graphics, and to extend this capability with high-performance interactivity over very large or streaming datasets. Bokeh helps to quickly and easily create interactive plots, dashboards, and data applications. It is built on top of HTML5 canvas, and designed for high-performance rendering of larger data sets.

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
output_notebook()

p = figure(plot_width = 450, plot_height = 350)
p.square(x=[2, 4, 6, 8], y=[4, 3, 7, 1], size = 16, color='blue', alpha = 0.3)  # We get squares for scatter plot
show(p)

# Change the markers - Diamond, square and circle
x = figure(plot_width = 450, plot_height = 350)
x.diamond(x=[2, 4, 6, 8], y=[4, 3, 7, 1], size = 14, color='blue', alpha = 0.3)
x.circle(x=[3, 8, 5, 9], y=[8, 4, 5, 3], size = 14, color='green', alpha = 0.3)
x.square(x=[1, 3, 2, 7], y=[1.5, 4.0, 5.7, 3.9], size = 14, color='red', fill_color='yellow', alpha = 0.9)
show(x)
```
